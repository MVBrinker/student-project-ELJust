# SIMPSON PARADOX


Simpson's paradox is characterized by a reversal in association. When two variables, say race and admission, are related to a third variable, then the trend that appears in the different groups of data disappears or reverses when these groups are combined.

Coming back to Simpson's paradox: when data is aggregated and two (categorical) variables are related to a third, a reversal of trends may appear.
